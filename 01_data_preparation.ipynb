```python
import pandas as pd
import pyodbc
from sqlalchemy import create_engine
import urllib

file_path = r"C:\Users\husse\Downloads\projects to perform\archive (1)\Supply chain logisitcs problem.xlsx"

sheets = [
    "OrderList",
    "FreightRates",
    "WhCosts",
    "WhCapacities",
    "ProductsPerPlant",
    "VmiCustomers",
    "PlantPorts"
]
dfs = {}
for sheet in sheets:
    df = pd.read_excel(file_path, sheet_name=sheet)
    dfs[sheet] = df
    print(f"{sheet} loaded: {df.shape[0]} rows, {df.shape[1]} columns")
params = urllib.parse.quote_plus(
    "DRIVER={ODBC Driver 17 for SQL Server};"
    "SERVER=localhost;"      
    "DATABASE=supply_chain;" 
    "Trusted_Connection=yes;"
)
engine = create_engine("mssql+pyodbc:///?odbc_connect=%s" % params)


for sheet, df in dfs.items():
    df.to_sql(sheet, engine, if_exists="replace", index=False)
    print(f"{sheet} uploaded to SQL Server")

query = "SELECT TOP 5 * FROM OrderList;"
df_check = pd.read_sql(query, engine)
print("First 5 rows from OrderList:")
print(df_check)
```
